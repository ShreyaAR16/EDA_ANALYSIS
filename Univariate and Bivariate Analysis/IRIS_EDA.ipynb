```python
#Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='class',data = df)
plt.show()
```

```python
#Comparing Sepal Length and Sepal Width
sns.scatterplot(x='sepallength', y='sepalwidth', hue='class', data=df)
plt.legend(bbox_to_anchor=(1,1), loc=2)
plt.show()
```

From the above plot, we can infer that  

Species Setosa has smaller petal lengths and widths.
Versicolor Species lies in the middle of the other two species in terms of petal length and width
Species Virginica has the largest of petal lengths and widths.

```python
#Multivariate Analysis
sns.pairplot(df.drop([],axis=1),hue='class',height=2)
```

Setosa has the smallest of petals widths and lengths. It also has the smallest sepal length but larger sepal widths.

```python
#Histogram
fig,axes = plt.subplots(2,2, figsize=(15,15))
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(data = df, x = 'sepallength',bins=6)
axes[0,1].set_title("Petal Length")
axes[0,1].hist(data = df, x = 'petallength',bins=6)
axes[1,0].set_title("Sepal Width")
axes[1,0].hist(data = df, x = 'sepalwidth',bins=6)
axes[1,1].set_title("Petal Length")
axes[1,1].hist(data = df, x = 'petalwidth',bins=6)

```

```python
#Distplot
plot = sns.FacetGrid(df, hue ='class')
plot.map(sns.distplot, "sepallength").add_legend;

plot = sns.FacetGrid(df, hue ='class')
plot.map(sns.distplot, "petallength").add_legend;

plot = sns.FacetGrid(df, hue ='class')
plot.map(sns.distplot, "sepalwidth").add_legend;

plot = sns.FacetGrid(df, hue ='class')
plot.map(sns.distplot, "petalwidth").add_legend;

plt.show()
```

```python
#Boxplot
def graph(y):
    sns.boxplot(x="class",y=y,data = df)
plt.figure(figsize=(10,10))
plt.subplot(221)
graph('sepallength')
 
plt.subplot(222)
graph('sepalwidth')
 
plt.subplot(223)
graph('petallength')
 
plt.subplot(224)
graph('petalwidth')
```

From the above graph, we can see that  

Species Setosa has the smallest features and less distributed with some outliers.
Species Versicolor has the average features.
Species Virginica has the highest features

```python
#Handling Outliers
sns.boxplot(x='sepalwidth', data=df)
```

```python
Q1 = np.percentile(df['sepalwidth'], 25,
                interpolation = 'midpoint')
 
Q3 = np.percentile(df['sepalwidth'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
print("Old Shape: ", df.shape)
print("Old Shape: ", df.shape)
 
# Upper bound
upper = np.where(df['sepalwidth'] >= (Q3+1.5*IQR))
 
# Lower bound
lower = np.where(df['sepalwidth'] <= (Q1-1.5*IQR))
 
# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
 
print("New Shape: ", df.shape)
 
sns.boxplot(x='sepalwidth', data=df)
```
